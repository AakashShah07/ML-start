```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
sns.displot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] =  pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]), columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(dataset["ApplicantIncome"], kde=True)  # kde=True adds a density curve


plt.subplot(1,2,2)
plt.title("After")
sns.histplot(dataset["ApplicantIncome_ss"], kde=True)


plt.show()
```
